````python


# Create a new column with the major's category
def categorize_major(major):
    if "computer" in major.lower():
        return "Computer Science Engineering"
    elif "electronics" in major.lower():
        return "Electronics Engineering"
    elif "mechanical" in major.lower():
        return "Mechanical Engineering"
    elif "civil" in major.lower():
        return "Civil Engineering"
    elif "chemical" in major.lower():
        return "Chemical Engineering"
    else:
        return "Other Engineering"

df["major_category"] = df['Major'].apply(categorize_major)



````

## 3. ***Research Questions***

````python

# Filter the data to only include individuals with a Computer Science Engineering degree
cs_data = df[df["major_category"] == "Computer Science Engineering"]

# Filter the data to only include individuals with one of the specified job titles
job_titles = ["Programming Analyst", "Software Engineer", "Hardware Engineer", "Associate Engineer"]
job_data = cs_data[cs_data["JobTitle"].isin(job_titles)]

# Calculate the mean and median salaries for these job titles
mean_salary = df["AnnualSalary"].mean()
median_salary = df["AnnualSalary"].median()

# Compare the mean and median salaries to the range mentioned in the article
if mean_salary >= 250000 and mean_salary <= 300000:
    print("The claim in the Times of India article is consistent with the data (based on mean salary).")
else:
    print("The claim in the Times of India article is not consistent with the data (based on mean salary).")

if median_salary >= 250000 and median_salary <= 300000:
    print("The claim in the Times of India article is consistent with the data (based on median salary).")
else:
    print("The claim in the Times of India article is not consistent with the data (based on median salary).")
````

````python
pivot = df.pivot_table(index='Major', columns='Sex', values='ID', aggfunc='count')

# Create a stacked bar chart
pivot.plot(kind='bar', stacked=True, figsize=(15,7))
plt.xlabel('Specialisation')
plt.ylabel('Count')
plt.title('Gender-Specialization Relationship')
plt.show()
````

````python
pivot = df.pivot_table(index='major_category', columns='Sex', values='ID', aggfunc='count')

# Create a stacked bar chart
pivot.plot(kind='bar', stacked=True, figsize=(15,7))
plt.xlabel('Specialisation')
plt.ylabel('Count')
plt.title('Gender-Specialization Relationship')
plt.show()
````

````python
from scipy.stats import chi2_contingency
# Create a contingency table of the counts of males and females by specialization
contingency_table = pd.crosstab(df['Major'], df['Sex'])

# Print the contingency table
print(contingency_table)

# Run a chi-square test for independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results of the chi-square test
print('Chi-square statistic: {:.2f}'.format(chi2))
print('p-value: {:.4f}'.format(p_value))
print('Degrees of freedom: {}'.format(dof))
print('Expected counts:\n', expected)

# Check if the p-value is less than 0.05, indicating a statistically significant association
if p_value < 0.05:
    print('There is a statistically significant association between gender and specialization.')
else:
    print('There is not a statistically significant association between gender and specialization.')
````

##   Removing Outliers 

````python
df['AnnualSalary'].unique()
````

````python
# Calculate the interquartile range (IQR)
Q1 = df['AnnualSalary'].quantile(0.25)
Q3 = df['AnnualSalary'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Replace outliers with NaN
df['AnnualSalary'] = np.where((df['AnnualSalary'] > upper_bound) | (df['AnnualSalary'] < lower_bound), np.nan, df['AnnualSalary'])

# Fill missing values with the median
median = df['AnnualSalary'].median()
df['AnnualSalary'] = df['AnnualSalary'].fillna(median)
````
